```python
import tensorflow as tf
import numpy as np
from PIL import Image
from pathlib import Path

# Load your TFLite model (replace with your actual path if different)
interpreter = tf.lite.Interpreter(model_path="../../mobile-models/disease_detection.tflite")
interpreter.allocate_tensors()

# Get input/output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Sample image from your dataset (pick any one)
sample_img_path = Path("../../datasets/new-plant-diseases/train/Apple___Apple_scab/0001.jpg")  # Change to any image

# Preprocess image (same as training)
img = Image.open(sample_img_path).resize((224, 224))
img_array = np.array(img, dtype=np.float32) / 255.0
img_array = np.expand_dims(img_array, axis=0)

# Run inference
interpreter.set_tensor(input_details[0]['index'], img_array)
interpreter.invoke()
predictions = interpreter.get_tensor(output_details[0]['index'])[0]

# Get class names (from your 31 classes â€“ adjust if needed)
class_names = sorted([d.replace('___', ' - ') for d in Path("../../datasets/new-plant-diseases/train").iterdir() if d.is_dir()])
top_3_idx = np.argsort(predictions)[-3:][::-1]
top_3 = [(class_names[i], predictions[i]) for i in top_3_idx]

print("Sample image prediction (Apple scab):")
for name, prob in top_3:
    print(f"{name}: {prob*100:.1f}%")
```
