```python
import os
import matplotlib.pyplot as plt
import random
from pathlib import Path

# Correct path (works from ml-service/notebooks/)
dataset_path = Path("../../datasets/new-plant-diseases/train")

classes = sorted(os.listdir(dataset_path))
print(f"Total classes: {len(classes)}")
print("Classes:", classes)

# Show 9 random sample images
fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.ravel()

for i in range(9):
    cls = random.choice(classes)
    img_file = random.choice(os.listdir(dataset_path / cls))
    img_path = dataset_path / cls / img_file
    
    img = plt.imread(img_path)
    axes[i].imshow(img)
    axes[i].set_title(cls.replace("___", " - ").replace("_", " "), fontsize=12)
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
